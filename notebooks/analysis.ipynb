```python
import pandas as pd
import numpy as np
```

```python
df_poll = pd.read_csv('../data/processed/df_processed.csv')
```

```python
df_poll.replace(' United Kingdom', 'United_Kingdom', inplace=True)
```

### Challenge 1

```python
df_ch1 = df_poll[['country', 'job_title', 'gender', 'uuid']]
```

```python
df_ch1 = df_ch1.groupby(['country', 'job_title', 'gender'])['uuid'].count().reset_index()
```

```python
df_ch1.rename(columns={'country':'Country', 'job_title':'Job title', 'gender':'Gender', 'uuid':'Quantity'}, inplace=True)
```

```python
df_ch1.replace(' United Kingdom', 'United_Kingdom', inplace=True)
```

```python
df_ch1[df_ch1['Country']=='United_Kingdom']
```

```python
country_filter = df_ch1['Country']=='United_Kingdom'
```

```python
df_country = df_ch1.loc[country_filter]
```

```python
df_country['Country percentage ‰'] = df_country['Quantity']/df_country['Quantity'].sum()*1000
```

```python
df_country
```

```python
df_country.to_json('../data/results/country_analysis.json')
df_country.to_csv('../data/results/country_analysis.csv')
```

### Challenge 2

```python
df_poll['basic_income_vote'].unique()
```

```python
df_opinion = df_poll[['basic_income_vote', 'basic_income_arguments_for', 'basic_income_arguments_against', 'uuid']]
df_opinion
```

```python
df_opinion['number_pro_args'] = df_opinion['basic_income_arguments_for'].apply(lambda x: len(x.split('|')))
df_opinion['number_con_args'] = df_opinion['basic_income_arguments_against'].apply(lambda x: len(x.split('|')))

# df_opinion = df_opinion[['basic_income_vote', 'number_pro_args', 'number_con_args']]

df_opinion
```

```python
df_opinion = df_opinion.groupby('basic_income_vote').agg(number_pro_args = ('number_pro_args','sum'),
                                            number_con_args = ('number_con_args','sum'),
                                            number_votes =('uuid', 'count')
                                           )
```

```python
df_opinion = df_opinion[['number_votes', 'number_pro_args', 'number_con_args']]
```

```python
df_opinion.reset_index(inplace=True)
```

```python
df_opinion.rename(columns= {'basic_income_vote': 'Vote_intention', 'number_votes':'Number_of_votes', 'number_pro_args': 'Number_pro_arguments', 'number_con_args': 'Number_con_arguments'}, inplace=True)
df_opinion.set_index('Vote_intention')
```

```python
df_opinion.to_json('../data/results/opinion_analysis.json')
df_opinion.to_csv('../data/results/opinion_analysis.csv')
```

```python
print('Starting top jobs/education level analysis.')

df_edu = df_poll[['full_time_job', 'education_level', 'job_title', 'uuid']]
df_edu = df_edu.groupby(['education_level', 'job_title']).agg(total=('uuid', 'count'))
df_edu = df_edu.reset_index()
df_edu = df_edu.groupby('education_level').apply(lambda x: x.nlargest(3, 'total')).reset_index(drop=True)
df_edu.rename(columns={'education_level': 'Education_level', 'job_title': 'Job_title','total': 'Total'}, inplace=True)
df_edu.set_index('Education_level', inplace=True)
df_edu.to_csv('../data/results/edu_level_analysis.csv')
df_edu
```
