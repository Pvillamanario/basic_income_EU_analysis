# Data analysis:

```python
import pandas as pd
import numpy as np
```

```python
df_poll = pd.read_json('../data/processed/df_processed.json')
```

### Challenge 1

```python
df_ch1 = df_poll[['country', 'job_title', 'gender', 'uuid']]
```

```python
df_ch1 = df_ch1.groupby(['country', 'job_title', 'gender'])['uuid'].count().reset_index()
```

```python
df_ch1.rename(columns={'country':'Country', 'job_title':'Job title', 'gender':'Gender', 'uuid':'Quantity'}, inplace=True)
```

```python
country_filter = df_ch1['Country']=='Spain'
```

```python
df_country = df_ch1.loc[country_filter]
```

```python
df_country['Country percentage ‰'] = df_country['Quantity']/df_country['Quantity'].sum()*1000
```

### Challenge 2

```python
df_poll['basic_income_vote'].unique()
```

```python
df_opinion = df_poll[['basic_income_vote', 'basic_income_arguments_for', 'basic_income_arguments_against', 'uuid']]
df_opinion
```

```python
df_opinion['number_pro_args'] = df_opinion['basic_income_arguments_for'].apply(lambda x: len(x.split('|')))
df_opinion['number_con_args'] = df_opinion['basic_income_arguments_against'].apply(lambda x: len(x.split('|')))

# df_opinion = df_opinion[['basic_income_vote', 'number_pro_args', 'number_con_args']]

df_opinion
```

```python
df_opinion.groupby('basic_income_vote').agg(number_pro_args = ('number_pro_args','sum'),
                                            number_con_args = ('number_con_args','sum'),
                                            number_votes =('uuid', 'count')
                                           )
```

```python
df_edu = df_poll[['full_time_job', 'education_level', 'job_title', 'uuid']]
df_edu
```

```python
working_filter = df_edu['full_time_job'] == 'yes'
```

```python
x = df_edu[working_filter].groupby(['education_level', 'job_title']).agg(total = ('uuid', 'count'))
x
```

```python
x.reset_index
```
